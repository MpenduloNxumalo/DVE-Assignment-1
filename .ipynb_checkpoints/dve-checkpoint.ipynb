```python
#this function checks if a string is a valid date and converts entries to datetime
def is_valid_date(date):
    if len(str(date))<10:
        return np.datetime64('NaT')
    res = True
    try:
        res = bool(parser.parse(date))
    except ValueError:
        res = False

    if res==True:
        list= date.split(' ')
        if len(list) == 3:
            new_date=list[2]+"/"+list[0]+"/"+list[1]
            if len(list[0]) != 3:
                new_date=list[2]+"/"+list[1]+"/"+list[0]
            date_time_obj = datetime.strptime(new_date, '%Y/%b/%d')
            return date_time_obj
    return np.datetime64('NaT')
    

print(is_valid_date("M"))
#df["date_fueled"]=df["date_fueled"].apply(is_valid_date)
#df['date_fueled'] = df.apply(lambda x: x['date_captured'] if (is_valid_date(x['date_captured'])!="NaD" and x["date_fueled"]=="NaD")else x['date_fueled'], axis=1)
```

## feature enginering

```python
def car(url):
    url=str(url)
    url=url[28:]
    list=url.split("/")    
    if (len(list)!=5):
        return list[0]
    else:
        return list[1]

def model(url):
    url=str(url)
    url=url[28:]
    list=url.split("/")
    if(len(list) == 3):
        return None
    if (len(list)==4):
        return list[1]
    else:
        return list[2]
                   
def year(url):
    url=str(url)
    url=url[28:]
    list=url.split("/")
    if(len(list) == 3):
        return list[1]
    if (len(list)==4):
        return list[2]
    else:
        return list[3]
                   
def userID(url):
    url=str(url)
    url=url[28:]
    list=url.split("/")
    if(len(list) == 3):
        return list[2]
    if (len(list)==4):
        return list[3]
    else:
        return list[4]
        
    
```

```python
df["car_make"]=df["user_url"].apply(car)
```

```python
df["model"]=df["user_url"].apply(model)
```

```python
df["year"]=df["user_url"].apply(year)
```

```python
df["UserID"]=df["user_url"].apply(userID)
```

```python
def currency(cost):
    cost=str(cost)
    for i in range(len(cost)):
        if cost[i].isnumeric():
            return cost[:i]
    return None

def float_val(cost):
    cost=str(cost)
    for i in range(len(cost)):
        if cost[i].isnumeric():
            return float(to_num(cost[i:]))
    return np.nan

float_val("$5.11")
```

```python
df["currency"]=df["total_spent"].apply(currency)
```
